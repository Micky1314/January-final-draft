```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\Filim\\Documents\\Jupyter Coursework Project\\Data\\HealthData.csv")

# Asking user for the option they want:

print("Welcome to the patient's records! Please read through the options and type in the number that you need:")
print("1 to see the number of people in each BMI category")
print("2 to see a particular patient's BMI")
print("3 to see which patients and the total number of hypoglycemic individuals")
print("4 to see which patients and the total number of hyperglycemic individuals")


chosen_answer = int(input("Which option would you like to choose?"))

if chosen_answer == 1:
    # 1: BMI checker
    
    # 1.1: Grouping continuous BMI by its clinical category
    
    # Tally per weight category
    underweight = 0
    healthy_weight = 0
    overweight = 0
    obese = 0
    severely_obese = 0
    
    # Checking which patients fall into which category
    for bmi in df["BMI"]:
        if bmi < 18.5:
            underweight += 1
        elif bmi >= 18.5 and bmi < 25:
            healthy_weight += 1
        elif bmi >= 25 and bmi < 30:
            overweight += 1
        elif bmi >= 30 and bmi < 40:
            obese += 1
        else:
            bmi >= 40
            severely_obese += 1
    
    # Print out the totals
    print(f"Number of individuals who are: \nUnderweight: {underweight} \nHealthy weight: {healthy_weight} \nOverweight: {overweight} \nObese: {obese} \nSeverely obese {severely_obese}")

if chosen_answer == 2:
    # 1.2: Finding a specific patient's BMI
    
    # Asking for the patient's id number
    inquiry = int(input("What is the patient ID?"))
    
    # Filter for subsetting by column
    bmi_columns = df["BMI"]
    
    # Searching within the id column and reporting the patient's BMI number
    for id in df["id"]:
        if id == inquiry:
            print(bmi_columns[inquiry])

# If diabetic, then examine whether they are also at risk from other conditions

# Blood glucose count
hypoglycemia = 0
hyperglycemia = 0

# Checking for who (and total) number of hypoglycemic individuals
if chosen_answer == 3:
    for index, row in df.iterrows():
        if row["diabetes"] == 1:
            if row["glucose"] < 60:
                hypoglycemia += 1
                print(f"Id number: {row["id"]}")
    print(f"Total number of hypoglycemic patients: {hypoglycemia}")

# Checking for who (and total) number of hyperglycemic individuals
if chosen_answer == 4:
    for index, row in df.iterrows():
        if row["diabetes"] == 1:
            if row["glucose"] > 150:
                hyperglycemia += 1
                print(f"Id number: {row["id"]}")
    print(f"Total number of hyperglycemic patients: {hyperglycemia}")
```

```python
# 1: Patient CHD risk calculator
def patient_risk(id):

    # Base risk by randomness
    risk_factor = 0.1
    
    patients_record = df.iloc[id]

    if patients_record["sex"] == "M":
        risk_factor += 0.1
    if patients_record["age"] > 40:
        risk_factor += 0.1
    if patients_record["is_smoking"] == "YES":
        risk_factor += 0.2
    if patients_record["prevalentStroke"] == 1:
        risk_factor += 0.05
    if patients_record["prevalentHyp"] == 1:
        risk_factor += 0.15
    if patients_record["diabetes"] == 1:
        risk_factor += 0.1
    if patients_record["totChol"] > 240:
        risk_factor += 0.15
    if patients_record["sysBP"] > 140:
        risk_factor += 0.05
    if patients_record["diaBP"] > 90:
        risk_factor += 0.05
    if patients_record["BMI"] >= 25:
        risk_factor += 0.1
    if patients_record["heartRate"] > 100:
        risk_factor += 0.1
    
    return round(risk_factor, 2)

# 2: Function for lifestyle suggestions:

def lifestyle_suggestion(id):
    
    patients_record = df.iloc[id]

    if patients_record["age"] > 40:
        print("You should maintain NHS lifestyle advice!")
        if patients_record["sex"] == "M":
            print("Regularly have a prostate checkup!")
        elif patients_record["sex"] == "F":
            print("Have a mammography every 1 to 2 years!")
    if patients_record["is_smoking"] == "YES":
        print("You should quit smoking!")
    if patients_record["diabetes"] == 1:
        print("You should ensure you visit your GP no less than every 6 months!")
    if patients_record["totChol"] > 240:
        print("Reduce your saturated fat consumption!")
    if patients_record["sysBP"] > 140 or patients_record["diaBP"] > 90:
        print("Limit caffeine!")
    if patients_record["heartRate"] > 100:
        print("Limit alcohol consumption!")
    if patients_record["glucose"] > 150 or patients_record["glucose"] < 60:
        print("You should consider the mediterranean diet!")

# 3: # Function that creates a plot using matplotlib
def plot_scatter(ax, x, y, c, xlabel, ylabel, title="My plot"):

    fig, ax = plt.subplots()
    
    ax.scatter(x, y, c=c, s=10, alpha=0.7)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

    plt.show()
```

# Section 3: Data Handling with Pandas (16 marks)

```python
# 1: Removing unwanted column
df2 = df.drop("TenYearCHD", axis='columns')

# 2: Checking for missing data
df2.isna().sum().plot(kind="bar")
print(df2.isna().sum())

# 3: Deleting NaN in columns which account for less than 5% of total column count (in this case only glucose accounts for more than 5%)
print(len(df2) * 0.05)
dropNaFilter = ["education", "cigsPerDay", "BPMeds", "totChol", "BMI", "heartRate"]
df3 = df2.dropna(subset=dropNaFilter)
print(df3.isna().sum())

# 4: Dealing with missing values in glucose column for which NaN's account for more than 5% of total data by mapping on the median for greater realism
print(f"Median glucose number is: {df3["glucose"].median()}")
df3.loc[:, "glucose"] = df3.loc[:, "glucose"].fillna(78)

# 5: Checking if any patients were duplicated
print(df3["id"].duplicated().sum())

# 6: Aggregating to extract insights for sex and education for BMI as a general indicator of health:
sex_education_agg = df3.groupby(["sex", "education"])["BMI"].agg(["mean", "std", "min", "max", "quantile"])
print(sex_education_agg)

# 7: Aggregating based on smoking habits:
#cigarette_impact = df3.groupby("is_smoking")[["BMI", "heartRate", "totChol"]].agg(["mean", "std", "min", "max", "quantile"])
#print(cigarette_impact)

df3.pivot_table(values="BMI",index="is_smoking", columns="sex", aggfunc=["mean", "std", "min", "max", "quantile"])
```

```python
from matplotlib import pyplot as plt

# 1 Visualising the distribution of BMI

fig, ax = plt.subplots()

ax.hist(df3["BMI"], bins=20, color="orange", edgecolor="red", label="BMI distribution")

ax.set_xlabel("BMI score")
ax.set_ylabel("Count")
ax.set_title("Overall BMI distribution")

ax.legend()

fig.savefig('Histogram.png', dpi=300)

plt.show()

# 2: The relationship between BMI and common risk factors

fig, ax = plt.subplots(3, 2, figsize=(14,10))

# Graph 1
ax[0, 0].hist(df3["totChol"], bins=20, edgecolor="black")
ax[0, 0].set_title("Cholesterol distrubution")
ax[0, 0].set_ylabel("Count")

# Graph 2
ax[0, 1].hist(df3["glucose"], bins=20, edgecolor="black")
ax[0, 1].set_title("Glucose distrubution")
ax[0, 1].set_ylabel("Count")

# Graph 3
ax[1, 0].hist(df3["heartRate"], bins=20, edgecolor="black")
ax[1, 0].set_title("Heart rate distrubution")
ax[1, 0].set_ylabel("Count")

# Graph 4
ax[1, 1].hist(df3["diaBP"], bins=20, edgecolor="black")
ax[1, 1].set_title("Diastolic blood pressure distrubution")
ax[1, 1].set_ylabel("Count")

# Graph 5
ax[2, 0].hist(df3["sysBP"], bins=20, edgecolor="black")
ax[2, 0].set_title("Systolic blood pressure distrubution")
ax[2, 0].set_ylabel("Count")

# Graph 6
ax[2, 1].hist(df3["age"], bins=20, edgecolor="black")
ax[2, 1].set_title("Age distrubution")
ax[2, 1].set_ylabel("Count")

fig.savefig('Subplots.png', dpi=300)

plt.suptitle("Histogram distributions of continuous variables:")
plt.show()

# 3: Scatterlpot of the relationshpip between systolic and diastolic blood pressure:

fig, ax = plt.subplots()
ax.scatter(df3["sysBP"], df3["diaBP"], s=5, alpha=0.55)

ax.set_facecolor("whitesmoke")
plt.grid()
ax.set_xlabel("Systolic blood pressure (mmHg)")
ax.set_ylabel("Diastolic blood pressure (mmHg)")
ax.set_title("Relationship between systolic and diastolic blood pressure")

fig.savefig('Scatterplot.png', dpi=300)

plt.show()
```

```python
import tkinter as tk
from tkinter import ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import matplotlib

# Plotting function for age distribution
def age_distribution():
    x1 = df3["age"]

    fig, ax = plt.subplots()
    ax.hist(x1)

    plt.show()

# Plotting function for BMI distribution
def bmi_distribution():
    x2 = df3["BMI"]

    fig, ax = plt.subplots()
    ax.hist(x2, color="red")

    plt.show()

# Plotting function for cigarrettes smoked/ sex
def sex_count():
    x3 = df3["sex"]
    y1 = df3["cigsPerDay"]

    fig, ax = plt.subplots()
    ax.bar(x3, y1)

    plt.show()


# Tkinter main window

matplotlib.use("TkAgg")

# Main window for GUI
main = tk.Tk()
main.title("Patient information")
main.geometry("1000x600")
main.configure(bg="azure2")

welcome = tk.Label(text="Welcome to the interactive system!").pack()

button1 = ttk.Button(main, text="Show complete patient age distribution", command=age_distribution)
button1.pack(side=tk.BOTTOM, pady=20)

button2 = ttk.Button(main, text="Show BMI distributions", command=bmi_distribution)
button2.pack(side=tk.BOTTOM, pady=30)

button3 = ttk.Button(main, text="Show sex count", command=sex_count)
button3.pack(side=tk.BOTTOM, pady=30)

main.mainloop()
```
